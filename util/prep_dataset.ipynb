# Scrape data from the Internet to create dataset

```python
import requests
from bs4 import BeautifulSoup
import cloudscraper
import re

# a popular website to practice JLPT exams, contains many reading exercises and listening transcript
WEBSITE = "https://japanesetest4you.com/"
```

```python
scraper = cloudscraper.create_scraper()
READING_EXERCISE = "https://japanesetest4you.com/japanese-language-proficiency-test-jlpt-n5-reading-exercise-3/#more-513"
response = scraper.get(READING_EXERCISE)
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
print(soup)
```

```python
p =soup.find_all(id="post-513")
print(len(p))
print(p)
```
